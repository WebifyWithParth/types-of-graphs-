```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[6,7,9,2,3,2]
plt.plot(x,y)
plt.title('line plot')
plt.xlabel('X axis')
plt.ylabel('Yaxis')
plt.show()
```

```python
categories=['A','B','C']
values=[5,7,3]
plt.bar(categories ,values)
plt.title('bar chart')
plt.xlabel('categories')
plt.ylabel('values')
plt.show()
```

```python
import numpy as np
data=np.random.randn(1000)
plt.hist(data,bins=11)
plt.xlabel('data')
plt.ylabel('categories')
plt.title('graph')
plt.show()
```

```python
data=[1,2,3,4]
categories=[5,6,7,8]
plt.scatter(data,categories)
plt.xlabel('papa')
plt.ylabel('beta')
plt.title('family')
```

```python
x=[1,2,3,4]
labels=['A','B','C','D']
plt.pie(x,labels=labels,autopct='%1.1f%%')
plt.title('pie chart')
plt.show()
```
